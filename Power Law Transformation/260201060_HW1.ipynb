```python
import cv2 as cv 
import matplotlib.pyplot as plt
import numpy as np
from skimage import color
from skimage import io
def plot_image(image_1, image_2, title_1, title_2):
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(image_1,cmap="gray")
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    plt.imshow(image_2,cmap="gray")
    plt.title(title_2)
    plt.show()
url = "https://camo.githubusercontent.com/dca7d906b6d63aa87c128059842ac413e352aac85f6046c1d22afb2105ef500f/687474703a2f2f692e696d6775722e636f6d2f397475316d61422e6a7067"
#I got the picture from a user who had a dark theme problem during the game.(I found it on google.)
image = io.imread(url)
plt.figure(figsize=(7,7))
plt.imshow(image)  
plt.show()
image_gray = color.rgb2gray(image)
gamma = 0.35
new_image = np.power(image_gray,gamma)
plot_image(image_gray, new_image, "Original", "Image After Power Law Transformation")
```
